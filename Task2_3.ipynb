```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image = cv2.imread('images2/james-peacock-1yRgTUVykA0-unsplash.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
     
```

```python
original_height, original_width = image.shape[:2]
aspect_ratio = original_width / original_height

new_width = 740
new_height = int(new_width / aspect_ratio)

resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_AREA)

resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
plt.imshow(resized_image_rgb)
plt.show()
```

```python
output_path = "resized_image.jpg"
Image.fromarray(resized_image_rgb).save(output_path, format='JPEG')

image_size = len(open(output_path, 'rb').read())

print(f"Коэффициент отношения сторон: {aspect_ratio:.3f}")
print(f"Высота полученного изображения: {new_height}")
print(f"Размер сжатого изображения в байтах: {image_size}")
```
